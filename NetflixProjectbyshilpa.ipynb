In this project, two datasets were used to build a personalized movie recommendation system. The first dataset contained customerID, movieID, and customer movie ratings. From this data, I extracted all unique movie IDs and analyzed each customer's historical rating behaviour.
To generate personalized recommendations, I applied Singular Value Decomposition (SVD), a matrix-factorization technique commonly used in collaborative filtering. SVD decomposes the item interaction matrix into latent factors, helping uncover hidden patterns such as user preferences and movie characteristics. This enabled the model to accurately predict how a customer might rate unseen movies.
To further optimize recommendation quality, I performed percentile-based filtering to identify and prioritize the highest-rated movies across the dataset.
Additionally, an SVM-based evaluation framework(cross validation ) was used to fine-tune and validate recommendation outputs, ensuring the system delivered the most relevant movie suggestions based on each customer's past rating history.
Overall, this project demonstrated the use of SVD for collaborative filtering, enhanced by statistical ranking techniques and machine-learning validation, resulting in a reliable and effective movie recommendation engine similar to the approach used by platforms like Netflix.
