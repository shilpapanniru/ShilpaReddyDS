Weather data contains non-linear relationships, interactions between multiple variables (temperature, humidity, pressure, wind), and a fair amount of noise. Random Forest handles this effectively by combining multiple decision trees trained on different subsets of data and features. This ensemble approach improves prediction stability, reduces overfitting, and performs well even with limited or imperfect weather datasets, making it a strong and reliable choice for weather forecasting problems.

```python
#build random forest regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
rf=RandomForestRegressor(n_estimators=500,min_samples_split=2,max_depth=20, bootstrap=True,random_state=42)
rf.fit(X_train_scaled,y_train)
y_pred_rf=rf.predict(X_test_scaled)
print(r2_score(y_test,y_pred_rf))
print(mean_squared_error(y_test,y_pred_rf))
print(mean_absolute_error(y_test,y_pred_rf))
```

### Random Forest Hyperparameters for Tuning:

1.  **`n_estimators`**: The number of trees in the forest.
    *   **Effect**: A higher number generally leads to better performance but increases computation time. Too many might lead to diminishing returns or overfitting if not coupled with other regularization.
    *   **Typical Range**: `10` to `2000` (often in steps like `100, 200, 500, 1000`).

2.  **`max_features`**: The number of features to consider when looking for the best split.
    *   **Effect**: Controls the diversity of the trees. Lower values reduce overfitting but can increase bias. Higher values can lead to less diverse trees.
    *   **Typical Options**: `'auto'`, `'sqrt'`, `'log2'`, or an integer/float representing a fraction of features.

3.  **`max_depth`**: The maximum depth of the tree.
    *   **Effect**: Controls the complexity of the trees. Deeper trees can capture more specific patterns but are more prone to overfitting. Limiting depth helps reduce overfitting.
    *   **Typical Range**: `None` (unlimited), `5` to `30` (or more, depending on data complexity).

4.  **`min_samples_split`**: The minimum number of samples required to split an internal node.
    *   **Effect**: Controls how small a node can be before it's not split further. Higher values prevent overfitting but can lead to underfitting.
    *   **Typical Range**: `2` to `20` (or a percentage of the total samples).

5.  **`min_samples_leaf`**: The minimum number of samples required to be at a leaf node.
    *   **Effect**: Similar to `min_samples_split`, it helps prevent creating leaf nodes that are too specific to the training data.
    *   **Typical Range**: `1` to `20` (or a percentage).

6.  **`bootstrap`**: Whether bootstrap samples are used when building trees.
    *   **Effect**: If `True` (default), samples are drawn with replacement. If `False`, the entire dataset is used to build each tree. Setting to `False` can reduce variance but might increase overfitting.
    *   **Typical Options**: `True` or `False`.

7.  **`random_state`**: Controls the randomness of bootstrapping and feature sampling.
    *   **Effect**: Important for reproducibility. Setting an integer ensures that the same results are obtained every time the model is run with the same parameters.
    *   **Typical Options**: Any integer (e.g., `42`).

### How to Tune:

You can use techniques like Grid Search (`GridSearchCV`) or Randomized Search (`RandomizedSearchCV`) from `sklearn.model_selection` to systematically explore different combinations of these hyperparameters and find the best performing model. For larger datasets, consider using more efficient optimization algorithms like Bayesian Optimization.

Weather data often exhibits complex, non-linear patterns, interactions across multiple variables, and varying seasonal effects. XGBoost handles these effectively through gradient-boosted decision trees, which iteratively learn from previous errors to improve prediction accuracy. Its ability to manage missing values, reduce bias and variance, and focus on difficult-to-predict patterns makes XGBoost highly suitable for weather forecasting tasks, especially when high predictive performance is required.

```python
#build Xgboost regressor
from xgboost import XGBRegressor
xgb=XGBRegressor(n_estimators=100,max_depth=20)
xgb.fit(X_train_scaled,y_train)
y_pred=xgb.predict(X_test_scaled)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

To improve the XGBoost model performance find the optimum parameters to tune the model using grid search CV. Perform a grid search using GridSearchCV to find the best hyperparameters for the XGBoost model with cross-validation.



```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [5, 10, 15],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 0.9],
    'colsample_bytree': [0.7, 0.8, 0.9]
}

# Instantiate the XGBoost Regressor model
xgb = XGBRegressor(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,  # cross validation 5 times
    scoring='neg_mean_squared_error',
    verbose=2,  # Added verbose to see the progress
    n_jobs=-1   # Use all available cores
)

# Fit GridSearchCV to the training data
grid_search.fit(X_train_scaled, y_train)

print("Grid search completed. Best parameters found:")
print(grid_search.best_params_)
```

 Train a new XGBoost model using the best hyperparameters obtained from the GridSearchCV result to improve the model performance.

```python
best_params = grid_search.best_params_

# Instantiate a new XGBoost Regressor model with the best parameters
xgb_best = XGBRegressor(**best_params, random_state=42)

# Train the model on the scaled training data
xgb_best.fit(X_train_scaled, y_train)

print("XGBoost model trained with best hyperparameters.")
```

Model Evaluation After tuning the model using grid search cv


```python
y_pred_xgb_best = xgb_best.predict(X_test_scaled)

print(f"R2 Score (XGBoost best): {r2_score(y_test, y_pred_xgb_best):.4f}")
print(f"Mean Squared Error (XGBoost best): {mean_squared_error(y_test, y_pred_xgb_best):.4f}")
print(f"Mean Absolute Error (XGBoost best): {mean_absolute_error(y_test, y_pred_xgb_best):.4f}")
```

To visualize the performance of the tuned XGBoost model, I will plot the actual y_test values against the predicted y_pred_xgb_best values.

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual')
plt.plot(y_test.index, y_pred_xgb_best, label='Predicted')
plt.title('Actual vs. Predicted Mean Temperature (XGBoost)')
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.legend()
plt.show()
```
